# Methodology <a class="anchor" id="method"></a>

This project conducts sentiment analysis of Twitter data with regards to the COVID-19 vaccines developed by pharmaceutical companies. We selected Twitter as our social media platform since it focuses on keywords and hashtags and allows posting to a wider audience when compared to other social media platforms. This, in addition to the character limit on Twitter compels users to be vocal about their opinions in as concise a way as possible. 

It is important to note that the aim of this project is not to rate and categorize different vaccines but to understand the general sentiment of the public towards (or against) vaccines and quantify this sentiment. We therefore believe analyzing more than one vaccine company as a keyword can help us better understand public perception.  
 
All tweets with keywords and hashtags such as #covid, #covid-19, #antivaxxers, #covidvaccine were included in the study. Only tweets in English were included and were coded in Python to conduct vaccine related sentiment analysis.  
   
Considering the efficiency of data storing and sharing, as well as the reproductivity of work, we normalized original two datasets to better fit the objective of this project. Besides this, we used Amazon Relational Database Service (Amazon RDS) to establish our own database and only queried necessary information for our analysis.

The Pandas package and NLTK (Natural Language ToolKit) were used for sentiment analysis of calculating subjectivity and polarity of all tweets. The plotly package and matplotlib package are used collaboratively in generating comparison of vaccine adoption behaviours in different locations.

We would highlight the methology of data processing and analysis part in following separated sections as 'Database Establishmend and Data Import' and 'Data Analysis'.

# Conclusions <a class="anchor" id="conclusion"></a>

Based on the demographic analysis of vaccination data around the world, North America has the leading place in average vaccination rate per capita. And according to the twitter data sentiment analysis, generally people of North America held a slight positive attitude towards COVID-19 vaccines. However, we could still summarize the main reasons for showing reluctance in getting vaccinated into three categories as follows.  

### Lack of Knowledge: <a class="anchor" id="lack-know"></a>
Most people are not aware of the exact information of vaccine administering. For example,  which brands of COVID-19 vaccines have been approved by the government health authorities, how many doses people need to receive and what will be the timeline of these doses. Moreover, the general public is not aware about what efficacy scientifically means and are misinformed that vaccines are a permanent cure. These factors combined make them reluctant towards the idea of getting vaccinated.   

**How our analysis helps with next steps**  
To counter the above, government health authorities and the pharmaceutical companies could explore educating people and pointing them to the right resources. Some tactics include having trusted organizations such as the CDC or pharmaceutical companies being more transparent and defining scientific terms as part of their rollout process. In addition to this, investing in educational advertisements (such as print and digital media) or conducting free virtual workshops organized by expert doctors are promising in achieving our goal of an increased vaccination rate.  

### Government & Politicians: <a class="anchor" id="gov-poli"></a>
Unfortunately, COVID-19 vaccination has become an extremely hot political selling point. Various politicians have been tweeting about the vaccine rollout but the numbers are far from reality. This creates a doubt and mistrust. Some of the information shared by government officials too is not accurate. This adds further complications to an already complex vaccine rollout process. All this confusion can immensely contribute to a lack of trust amongst individuals and consequently resulting in them being reluctant to get vaccinated. The issue can be handled with well thought out government plans which cater to all population groups in an organized and need-based order. For countries dealing with economic turmoil, adherence to World Health Organization (WHO) guidelines and procuring vaccines have been nothing short of an adventure. Although financial institutions like the International Monetary Fund (IMF) and World Bank have helped certain countries, a lot needs to be done, on an individual level, to ensure vaccination rates continue to increase.  

**How our analysis helps with next steps**  
Our analysis can help governments and politicians collaborate with organizations such as the World Health Organization (WHO) and other public health organizations along with news agencies to limit the spread of misinformation while also taking care to not inflate statistics. This can contribute towards a rapid adoption of vaccinations.  


### Fear: <a class="anchor" id="fear"></a>
Fear is one of another bigger concerns for individuals who are not pro-vaccine. Since the rollout of AstraZeneca, several cases of blood clotting emerged from different parts of the world. This has made a lot of folks skeptical even for Pfizer and Moderna vaccines which have reportedly high efficacy rates of 95%. Many fear of potential side-effects would lead to an extra health cost, especially in countries like the USA where the health system is not centralized and subsidized by the government. Besides this, the fear of being re-infected despite having two doses of vaccine, also exists among people. Overcoming the fear is probably the biggest challenge.  

**How our analysis helps with next steps**  
As we are seeing an increase in the number of people getting vaccinated, especially in North America, knowing people‚Äôs biggest concern can be used to persuade those who are afraid of possible side effects. The BioTech companies can share more proof of the efficacy and safety of vaccines to eliminate public fear.  
By promoting the longer run benefits of these vaccines and emphasizing on how their benefits outweigh costs, the general vaccination adoption rates can be increased.  

### Concluding statements: <a class="anchor" id="conclu"></a>
Our study aims to answer the question: Can the rate of vaccinations be increased?
The above sentiment analysis not only helps us identify and classify major public concerns into broader categories but also gives a sense of future direction for the North American governments and pharmaceutical companies. ### Data Normalization <a class="anchor" id="data-norm"></a>

```python
# Normalization part: public schema
load_dotenv()

conString = {'host':os.environ.get('DB_HOST_F'),
             'dbname':os.environ.get('DB_NAME_F'),
             'user':os.environ.get('DB_USER_F'),
             'password':os.environ.get('DB_PASS_F'),
             'port':os.environ.get('DB_PORT_F')}

conn = psycopg2.connect(**conString)

query = """
create table if not exists public.users(
user_id bigint primary key,
user_name text not null,
user_verified boolean not null,
user_location text not null
);
create table if not exists public.tweets(
tweet_id bigint primary key,
user_id bigint REFERENCES public.users(user_id),
date date not null,
text text not null,
hashtags text not null
);
create table if not exists public.locations(
iso_code text primary key,
location text not null
);
create table if not exists public.vaccines(
iso_code text REFERENCES public.locations(iso_code),
date date not null,
people_vaccinated integer,
people_vaccinated_per_hundred decimal,
primary key (iso_code, date)
);"""

conn.rollback()
cur = conn.cursor()
cur.execute(query)
conn.commit()
```

### Data Import into Database <a class="anchor" id="data-import"></a>

```python
# read dataset
vaccine = pd.read_csv("vaccinations.csv")

# prepare sub-df for location table
location_df= vaccine.loc[:, ['location', 'iso_code']]
location_df = pd.DataFrame.drop_duplicates(location_df)

# create tuple list for inserting
location_clean = [tuple([v.get("location"), 
                          v.get("iso_code")])
                          for v in location_df.to_dict('records')]

# import data into location table
cur = conn.cursor()
conn.rollback()
psycopg2.extras.execute_batch(cur, """INSERT INTO public.locations
                        (location, iso_code) 
                    VALUES (%s, %s);""", 
                    location_clean)
conn.commit()
```

```python
# prepare sub-df for vaccine table
vacc_df= vaccine.loc[:, ['iso_code', 'date', 'people_vaccinated', 'people_vaccinated_per_hundred']]
vacc_df = pd.DataFrame.dropna(vacc_df)

# create tuple list for inserting
vacc = [tuple([v.get('iso_code'), 
               v.get('date'),
               v.get('people_vaccinated'),
               v.get('people_vaccinated_per_hundred')]) for v in vacc_df.to_dict('records')]

# import data into location table
cur = conn.cursor()
conn.rollback()
psycopg2.extras.execute_batch(cur, """INSERT INTO public.vaccines
                        (iso_code, date, people_vaccinated, people_vaccinated_per_hundred) 
                    VALUES (%s, %s, %s, %s);""", 
                    vacc)
conn.commit()
```

```python
# user pre-cleaned tweet_new for users table
user_df = tweet_new.loc[:, ['user_name', 'user_verified', 'user_location']]
user_df = user_df.drop_duplicates(subset = 'user_name', keep = 'first')

# create a random number list for user_id
user_df.loc[:, 'user_id'] = random.sample(range(999999999), len(user_df))

# create tuple list for inserting
user = [tuple([v.get('user_id'), 
               v.get('user_name'),
               v.get('user_verified'),
               v.get('user_location')]) for v in user_df.to_dict('records')]

# import data into location table
cur = conn.cursor()
conn.rollback()
psycopg2.extras.execute_batch(cur, """INSERT INTO public.users
                        (user_id, user_name, user_verified, user_location) 
                    VALUES (%s, %s, %s, %s);""", 
                    user)
conn.commit()
```

```python
# user pre-cleaned tweet_new for tweets table
tweet1 = tweet_new.loc[:, ['user_name','id', 'date', 'text', 'hashtags']]
tweet2 = tweet1.drop_duplicates(subset = 'id', keep = 'first')

# merge with user_df to get user_id
df_merge = tweet2.merge(user_df, how = 'left', on = 'user_name')
tweet_df = df_merge.loc[:, ['id','user_id','date', 'text', 'hashtags']]
tweet_df = tweet_df.rename(columns = {'id':'tweet_id'})

# create tuple list for inserting
tweet = [tuple([v.get('tweet_id'), 
                v.get('user_id'),
                v.get('date'),
                v.get('text'),
                v.get('hashtags')]) for v in tweet_df.to_dict('records')]

# import data into location table
cur = conn.cursor()
conn.rollback()
psycopg2.extras.execute_batch(cur, """INSERT INTO public.tweets
                        (tweet_id, user_id, date, text,hashtags) 
                    VALUES (%s, %s, %s, %s, %s);""", 
                    tweet)
conn.commit()
```

### Data Import for Analysis <a class="anchor" id="data-import2"></a>
Retrieve all the required data for the analysis. Include brief data descriptions, and DOIs where possible.

```python
# query vaccine data with locations
conn = psycopg2.connect(**conString)
cur = conn.cursor()
conn.rollback()

vacc_data = """
select v.date, v.people_vaccinated, v.people_vaccinated_per_hundred, l.location
from public.locations as l
inner join public.vaccines as v on v.iso_code = l.iso_code  
where l.iso_code like 'OWID%'
"""
cur.execute(vacc_data)
vacc_df = pd.DataFrame(cur.fetchall(), columns = ['Date','People_vaccinated','People_vaccinated_per_hundred', 'Location'])
```

```python
# query tweet data from database
conn = psycopg2.connect(**conString)
cur = conn.cursor()
conn.rollback()

tweet_data = """
select text from public.tweets
"""
cur.execute(tweet_data)
tup_lst = cur.fetchall()
```

### Demographic Analysis of People Vaccinated <a class="anchor" id="demo-ana"></a>

```python
# slice df for only data for each continent and whole world
loc_list = ('Africa', 'Asia', 'Europe', 'Oceania', 'North America', 'South America','World')
vacc_fin = vacc_df[vacc_df['Location'].apply(lambda x: any(k in x for k in loc_list if x != 'European Union'))]
```

```python
# draw time series plot
fig = px.line(vacc_fin, 
              x='Date',
              y='People_vaccinated_per_hundred', 
              color = 'Location', 
              title='Time series of people vaccinated per hundred')
fig.show()
```

```python
# select latest people vaccinated for each place
people_by_continent = vacc_df['People_vaccinated_per_hundred'].groupby(vacc_fin['Location']).max()

df = pd.DataFrame(people_by_continent)

df['People_vaccinated_per_hundred'] = pd.to_numeric(df['People_vaccinated_per_hundred'])

df.sort_values(by=['People_vaccinated_per_hundred'], inplace=True)

label = df['People_vaccinated_per_hundred'].tolist()
```

### Sentiment Analysis of Tweets <a class="anchor" id="senti-ana"></a>

```python
# use tup_lst to create df for further analysis
tweet_new = pd.DataFrame(tup_lst)
tweet_new = tweet_new.rename(columns = {0:'text'})
```

```python
# define stopwords
stopwords = nltk.corpus.stopwords.words('english')
stopwords = stopwords + ['https', 'moderna', 'vaccine', 'covid19', 'pfizer', 'pfizerbiontech', 'ontario', 'toronto',
'vaccines', 'covid', 'covid19vaccine', 'covidvaccine']
```

```python
#sentiment analysis
#check some high polarity sentences
from textblob import TextBlob

for s in tweet_new['text']:
    tb = TextBlob(s)
    pol = tb.sentiment.polarity
    if abs(pol) > 0.4:
        print(pol)
        print(s)
        print()
```

```python
# filter negative polarity text
Neg_pola = [s for s in tweet_new['text'] if TextBlob(s).polarity < -0.4]
Neg_pola = pd.DataFrame(Neg_pola)
Neg_pola = Neg_pola.rename(columns = {0:'text'})
```

```python
#collect sentiment scores for each sentence
sub_list = [TextBlob(s).sentiment.subjectivity for s in tweet_new['text']]
pol_list = [TextBlob(s).sentiment.polarity for s in tweet_new['text']]
```

```python
# visualize the sentiment scores using histogram
plt.hist(sub_list, bins=10)
plt.suptitle('Subjectivity of Tweet Text')
plt.xlabel('subjectivity score')
plt.ylabel('sentence count')
plt.grid(True)
plt.show()
```

```python
# visualize the polarity scores using histogram
plt.hist(pol_list, bins=10)
plt.suptitle('Polarity of Tweet Text')
plt.xlabel('polarity score')
plt.ylabel('sentence count')
plt.grid(True)
plt.show()
```

```python
# convert to lower case
tweet_new['text'] = tweet_new['text'].str.lower()
# Tokenize
tweet_new['text']  = tweet_new['text'].apply(word_tokenize)
# Remove Stop words
tweet_new['text'] = tweet_new['text'].apply(lambda x: [word for word in x if word not in stopwords])
# Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
tweet_new['text'].apply(lambda x: [wordnet_lemmatizer.lemmatize(word) for word in x])
# remove symbols
tweet_new['text'] = tweet_new['text'].apply(lambda x: [t for t in x if t.isalnum()])
tweet_new = tweet_new.reset_index()
```

```python
# general word frequency
word_list = [tweet_new['text'][i] for i in range(len(tweet_new['text']))]
word_list = [item for sublist in word_list for item in sublist]

separator = ' '
text = [separator.join(tweet_new['text'].iloc[i]) for i in range(len(tweet_new['text']))]
```

```python
# create general key word list
from collections import Counter
top = Counter(word_list).most_common(20)
print(top)
```

```python
# create bar chart of general key words
top_df = pd.DataFrame(top)
top_df = top_df.rename(columns = {0:'Word', 1:'Count'})
plt.figure(num=None, constrained_layout=True, figsize=(12, 8), 
                 dpi=80, facecolor='w', edgecolor='k')
plt.bar(top_df['Word'], top_df['Count'], color = 'green')
plt.suptitle('Word Frequency Bar Chart', fontsize = 20)
plt.ylabel('word count', fontsize = 15)
plt.xticks(rotation=60)
plt.rcParams.update({'font.size': 12})
plt.show()
```

```python
# preparing wordcloud
text_str=''
for i in text:
        text_str = text_str + i + ' '
        
text_str[:50]
```

```python
# create wordcloud plot for general key words
wordcloud = WordCloud(width=1000, height=700, 
                      background_color ='white', 
                      min_font_size = 10).generate(text_str)
plt.figure(figsize=(20,10), facecolor = None)
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)

plt.show()
```
